```python
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
print("Колонки с отсутствующими значениями и их количество:\n", missing_values)
```

```python
#если строковый тип данных, заполняем модой, иначе (числовой) - медианой

for col in missing_values.index:
    if df[col].dtype == 'object':
        mode_value = df[col].dropna().mode()
        df[col] = df[col].fillna(mode_value[0])
        
    else:
        df[col] = df[col].fillna(df[col].median())  
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# гистограмма для SalePrice
plt.figure(figsize=(8, 6))
sns.histplot(df['SalePrice'], kde=True, bins=30)
plt.title('Распределение SalePrice')
plt.xlabel('SalePrice')
plt.show()
```

```python
# 1) Распредление смещено вправо: бОльшая часть значений сосредоточена в области низких цен; длинный хвост тянется к области высоких цен
# 2) Распределение ненормальное: оно не симметрично и имеет длинный хвост
# 3) Имеются выбросы: дома дороже 400000, которые значительно отдалены от основной массы данных. 
# Вероятно, это небольшое количество чень дорогих элитных домов.
```

```python
# Исследуйте взаимосвязи между целевой переменной и другими характеристиками. Есть ли сильные корреляции?
```

```python
# Корреляционная матрица
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()
top_corr_features = correlation_matrix['SalePrice'].sort_values(ascending=False).head(10)
print("Наиболее коррелирующие признаки:\n", top_corr_features)
```

```python
features = ['OverallQual', 'GrLivArea', 'GarageCars']
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[feature], y=df['SalePrice'])
    plt.title(f'Взаимосвязь {feature} и SalePrice')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
# Взаимосвязь между качеством дома (OverallQual) и его ценой продажи (SalePrice):
# 1) чем выше качество дома, тем выше цена его продажи (положительная корреляция)
# 2) низкие характеристики качества дома (0-4) дают цену, сконцентированную в области до 200.000
# 3) высокие характеристики качества дома (8-10) дают большой разброс цен - от 550.000 до > 700.000
# 4) есть выбросы: в категории качества 4 есть объект с необычно высокой ценой, а в категории качества 10 - имеются очень дешевые объекты
```

```python
# Взаимосвязь между общей жилой площадью дома (GrLivArea) и его ценой продажи (SalePrice):
# 1) чем выше общая жилая площадь дома, тем выше цена его продажи (положительная корреляция)
# 2) для значений жилой площади более 3000 наблюдается увеличение разброса цен, что может быть связано с различиями в других характеристиках домов
# 3) есть выбросы: дома с площадью 4000-5000, но низкой ценой
```

```python
# Взаимосвязь между количеством мест для автомобилей в гараже (GarageCars) и его ценой продажи (SalePrice):
# 1) чем больше количество мест в гараже, тем выше цена его продажи (положительная корреляция)
# 2) основную массу домов составляют дома с 3 местами в гараже
# 3) есть выбросы: дома с 4 местами в гараже имеют неожиданно низкую стоимость 
```

```python
# a. Работайте с категориальными переменными, применяя one-hot encoding или label encoding, в зависимости от характера переменной.
```

```python
# One-hot encoding для категориальных переменных
categorical_cols = df.select_dtypes(include='object').columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
#При необходимости создайте новые характеристики, такие как общая площадь или возраст объекта недвижимости, путем объединения существующих характеристик.
```

```python
# Создание новых признаков, например, общей площади
df['TotalArea'] = df['GrLivArea'] + df['TotalBsmtSF']

# Пример: добавление возраста дома
df['HouseAge'] = df['YrSold'] - df['YearBuilt']
```
